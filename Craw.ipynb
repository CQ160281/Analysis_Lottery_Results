```python
# Import libraries and packages
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from datetime import datetime, timedelta
import pandas as pd

# Khai báo biến browser
browser = webdriver.Chrome()

# Khai báo biến current_date thời gian hiện tại
current_date = datetime(2021,5,4)

# List data rỗng
data = []
index = 0

while True:
    print("Process 300 days from {}-{}-{}".format(current_date.day, current_date.month, current_date.year))

    # Mở trang web
    url = 'https://www.thantai.net/so-ket-qua'
    browser.get(url)

    # Set date - Biên độ ngày
    end = browser.find_element(By.ID, "end")
    end.clear()
    end.send_keys("{}-{}-{}".format(current_date.day, current_date.month, current_date.year))

    # Nút 300 lượt
    btn = browser.find_element(By.XPATH, "/html/body/div[2]/main/div/form/div[2]/div/button[9]")
    btn.click()

    # Tìm tất cả class (trong class thay dấu cách bằng dấu chấm)
    result = browser.find_elements(By.CLASS_NAME, "font-weight-bold.text-danger.col-12.d-block.p-1.m-0")
    
    
    for row in result:
        # In ra các row text để biết chương trình đang run
        print(row.text)
        index +=1 # Tăng index (biến đếm)
        data.append(row.text) # sau đó đưa vào list data

    current_date -= timedelta(days = 300)
    # sleep(1)
    if index > 20*365: # Sau khi tăng đếm lên > 20 năm thì thoát ra
        break

# Ghi vào file csv
df = pd.DataFrame(data, columns=['KQ'])
df.to_csv("XSMB.csv", index=False)
browser.close()
```

```python
# Import libraries and packages
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from datetime import datetime, timedelta
import pandas as pd

# Khai bo bin browser
browser = webdriver.Chrome()

# Khai bo bin current_date thi gian hin ti
current_date = datetime(2021,5,4)

# List data rng
data = []
index = 0

while True:
    print("Process 300 days from {}-{}-{}".format(current_date.day, current_date.month, current_date.year))

    # M trang web
    url = 'https://www.thantai.net/so-ket-qua'
    browser.get(url)

    # Set date - Bin  ngy
    end = browser.find_element(By.ID, "end")
    end.clear()
    end.send_keys("{}-{}-{}".format(current_date.day, current_date.month, current_date.year))

    # Nt 300 lt
    btn = browser.find_element(By.XPATH, "/html/body/div[2]/main/div/form/div[2]/div/button[9]")
    btn.click()

    # Tm tt c class (trong class thay du cch bng du chm)
    result = browser.find_elements(By.CLASS_NAME, "font-weight-bold.text-danger.col-12.d-block.p-1.m-0")
    
    
    for row in result:
        # In ra cc row text  bit chng trnh ang run
        print(row.text)
        index +=1 # Tng index (bin m)
        data.append(row.text) # sau  a vo list data

    current_date -= timedelta(days = 300)
    # sleep(1)
    if index > 20*365: # Sau khi tng m ln > 20 nm th thot ra
        break

# Ghi vo file csv
df = pd.DataFrame(data, columns=['KQ'])
df.to_csv("XSMB.csv", index=False)
browser.close()
```
